### Push Today's Files to Your GitHub Repo

Right now, today's files exist in the codespace's filesystem. If you were to delete this codespace, these files would disappear with it. To make sure you don't lose these files, you'll want to turn the color of these new files' filenames from green with a "U" to their right into black (with no letter to the right). When I say this, I'm looking at the IDE's file explorer panel.

> In the terminal, executing `git status` gives you similar information.

Pushing these new files to your GitHub repo will ensure that you don't lose them. This is a multistep process:

1.  You "stage changes" either by clicking the plus-sign (+) next to each filename in the Source Control panel, or by running `git add <filename>` in the terminal window.

2.  With the changes you want staged, you then put a message in the textbox above the blue "Commit" button and click the "Commit" button (when working in the Source Control panel), or by running `git commit -m "<commit_message>"` in the terminal window. This commits your new files to the codespace's copy of your GitHub repo, which unfortunately will still disappear if you delete the codespace.

> NOTE: If you leave the `-m` flag off the git-commit command, git will open an editor and allow you to type a commit message. Including a commit message is a Good Practice. Ask an AI chatbot how to change the default editor into one that you know how to use.

3.  So, the last step is to push what you've committed to your GitHub repo. You can either click the "Push" icon above the list of commit messages, or you can run `git push` in the terminal window. Your work now appears in your CS1066 GitHub repo, and the files will appear in any new codespace you create (because codespaces pull from your GitHub repo).

Back in the IDE's file browser panel, the filenames of the new files should be black, and there should be no letter to the right.

> In the terminal, running `git status` now should report that "Your branch is up to date with 'origin/main'" and that there's "nothing to commit, working tree clean."

Check `https://github.com/<your_username>/cs1066` now and you'll see that the new files are there.
